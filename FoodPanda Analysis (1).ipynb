```python
plt.figure(figsize = (10,5))
df["gender"].value_counts().plot(kind="bar", color=["orange", "skyblue", "pink"])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Gender Values")
plt.tight_layout()
plt.grid(axis="y")
plt.show()
```

```python
df["restaurant_name"].value_counts()
```

```python
plt.figure(figsize = (10,5))
df["restaurant_name"].value_counts().plot(kind = "bar", color = ["Pink","Purple","yellow","silver","Gold"])
plt.title("Top 5 Restaurant Name")
```

```python
plt.figure(figsize = (10,5))
df["city"].value_counts().head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Cities by Customers")
plt.show()
```

```python
plt.figure(figsize = (10,5))
df["order_date"] = pd.to_datetime(df["order_date"])
df.groupby(df["order_date"].dt.month)["order_id"].count().plot(kind="line", marker="o", color = "Red", linestyle = "--")
plt.title("Monthly Orders Trend")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.show()
```

```python
df["dish_name"].value_counts()
```

```python
plt.figure(figsize = (10,5))
df["dish_name"].value_counts().plot(kind = "bar", color = "gray")
plt.title("Top 5 Dish Name")
plt.xlabel("Dish Name")
plt.ylabel("Number of Orders")
plt.show()
```

```python
plt.figure(figsize = (10,5))
df["revenue"] = df["quantity"] * df["price"]
df.groupby("city")["revenue"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top 10 Cities by Revenue")
plt.show()
```

```python
avg_order_value = df["revenue"].sum() / df["order_id"].nunique()
print("Average Order Value:", avg_order_value)
```
